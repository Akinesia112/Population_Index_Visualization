### 直方圖（Histogram）

Html color codes: https://htmlcolorcodes.com/

```
arr_df = df['人口密度'].to_numpy()
fig = plt.figure(figsize=(8,5), dpi = 100)
ax = sns.distplot(arr_df, bins = 100, kde = True, color = '#F5B7B1') # Bins & KDE
ax.set_title('Population Density Histagram')

# 加減3個標準差
plt.axvline(x=np.mean(arr_df)-3 * np.std(arr_df), color='r', linestyle='-')
plt.axvline(x=np.mean(arr_df)+3 * np.std(arr_df), color='r', linestyle='-')
```

二維異常值偵測（Bivariate Outlier Detection）- 紅色橢圓型是信賴區間。

```
from sklearn.covariance import EllipticEnvelope
clf = EllipticEnvelope()

legend = {}
xx, yy = np.meshgrid(np.linspace(0, 4, 100), np.linspace(0, 100, 100))
plt.figure(1, figsize=(5,5))
clf.fit(df[['戶數','扶養比']].values)
Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
legend['EllipticEnvelope'] = plt.contour(
    xx, yy, Z, levels=[0], linewidths=2, colors=['r'])

legend_values_list = list(legend.values())
legend_keys_list = list(legend.keys())

plt.figure(1, figsize=(5,5))# two clusters
plt.title("Bivariate Outlier detection")
plt.scatter(df['戶數'].values, df['扶養比'].values, color='blue')
bbox_args = dict(boxstyle="round", fc="0.8")
arrow_args = dict(arrowstyle="->")

plt.xlim((xx.min(), xx.max()))
plt.ylim((yy.min(), yy.max()))
plt.legend(legend_values_list[0].collections, legend_keys_list,
           loc="upper center");
plt.xlabel("戶數"); 
plt.ylabel("扶養比")
```

相關矩陣（Correlation Matrix）

```
import altair as alt
```

```
df_corr = df[['男女比',	'戶數',	'人口密度',	'扶養比',
                 '老化指數']].corr().stack().reset_index().rename(columns={0: 'correlation', 'level_0': 'variable', 'level_1': 'variable2'})
df_corr['correlation_label'] = df_corr['correlation'].map('{:.2f}'.format)
df_corr
```

```
base = alt.Chart(df_corr).encode(
    x='variable2:O',
    y='variable:O'    
)

# Text layer with correlation labels
# Colors are for easier readability
text = base.mark_text().encode(
    text='correlation_label',
    color=alt.condition(
        alt.datum.correlation > 0.5, 
        alt.value('white'),
        alt.value('black')
    )
)

# The correlation heatmap itself
cor_plot = base.mark_rect().encode(
    color='correlation:Q'
).properties(
    width = 500,
    height = 500
)

cor_plot + text
```

```
alt.Chart(df).transform_window(
    index='count()'
).transform_fold(
    ['男女比',	'扶養比','戶數','老化指數']
).mark_line().encode(
    x='key:N',
    y='value:Q',
    color='人口密度:Q',
    detail='index:N',
    opacity=alt.value(0.7)
).properties(width=300)
```
